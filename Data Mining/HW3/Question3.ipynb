```python
import pandas as pd

# Question 3

df = pd.read_csv("smartphone.csv")
display(df)
```

```python
# 3_a

data_crosstab = pd.crosstab(df['Company'], df['Capacity']) 
display(data_crosstab)
```

```python
# 3_b

data_crosstab = pd.crosstab([df.Weight, df.inch], df['Company']) 
display(data_crosstab)
```

```python
# 3_c

crosstab_03 = pd.crosstab(df['OS'], df['Company'])
crosstab_03.plot(kind = 'bar', stacked = True)
```

```python
crosstab_norm3 = crosstab_03.div(crosstab_03.sum(1), axis = 0)
crosstab_norm3.plot(kind = 'bar', stacked = True)
```
