```python
from sklearn.datasets import load_boston
import pandas as pd

# Question 2
# 2_a
features, target = load_boston(return_X_y=True)
```

```python
# 2_b
features = pd.DataFrame(features)
display(features.head())
```

```python
# 2_c
features['Price'] = target
display(features.head())
```

```python
# 2_i
print(features.isna().sum())
```

```python
import matplotlib.pyplot as plt

i=0; fig, ax = plt.subplots(1, 14, sharex='col', sharey='row',figsize=(50,5))
for col in features.columns:
    features[col].plot(kind='hist',title=col , ax=ax[i])
    i = i + 1
```

```python
for col in features.columns:
    print('column : ', col) 
    print('min = ', features[col].min(), '\nmax = ', features[col].max())
    print('=========================================')
```

```python
from sklearn.preprocessing import normalize

features_normalize = pd.DataFrame(normalize(features),columns = features.columns)
display(features_normalize)
```

```python
import matplotlib.pyplot as plt

i=0; fig, ax = plt.subplots(1, 14, sharex='col', sharey='row',figsize=(50,5))
for col in features_normalize.columns:
    features_normalize[col].plot(kind='hist',title=col , ax=ax[i])
    i = i + 1
```

```python
for col in features_normalize.columns:
    print('column : ', col) 
    print('min = ', features_normalize[col].min(), '\nmax = ', features_normalize[col].max())
    print('=========================================')
```

```python
from sklearn.model_selection import train_test_split
# 2_j
x = features_normalize.drop(columns=['Price'])
y = features_normalize[['Price']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 5)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
# 2_d
reg = LinearRegression()
reg.fit(x_train, y_train)
y_predict = reg.predict(x_test)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
# 2_e
################### methods for calculation of MAE, MSE, RMSE ###################
print('methods for calculation of MAE, MSE, RMSE')
print('MAE  = ', mean_absolute_error(y_test, y_predict))
print('MSE  = ', mean_squared_error(y_test, y_predict))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_predict)))

################### manual calculation of MAE, MSE, RMSE ###################
print('\nmanual calculation of MAE, MSE, RMSE')
bound = y_test.shape[0]
MAE = 0; MSE = 0
for i in range(0, bound):
    MSE += ((y_test.values.ravel()[i] - y_predict[i]) ** 2)
    MAE += (abs(y_test.values.ravel()[i] - y_predict[i]))

print('MAE  = ', MAE/bound)
print('MSE  = ', MSE/bound)
print('RMSE = ', np.sqrt(MSE/bound))
```

```python
from sklearn.model_selection import cross_val_score
# 2_f
cvs = cross_val_score(reg, x_train, y_train, cv=5)
print(cvs)
print('mean = ', cvs.mean())
```
