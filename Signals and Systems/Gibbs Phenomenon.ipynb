```python
import numpy as np
import matplotlib.pyplot as plt

def Pulse_Generation(N):
    
    x_t = np.linspace(0, 5, 500)
    x_t_fourier_series = np.zeros_like(x_t)
    for k in range(1, N):
        x_t_fourier_series += np.sin((2*k-1) * np.pi * x_t) / (2*k-1)
        
    x_t_fourier_series *= (4 / np.pi)
    square_wave = 1 - 2 * (x_t.astype(int) % 2)
    fig, ax = plt.subplots(figsize=(15, 5))
    ax.plot(x_t, square_wave, lw=5, alpha=0.5, c='y')
    ax.plot(x_t, x_t_fourier_series, 'r')
    ax.set_ylim(-1.2, 1.2)

    ax.set_xticks([i for i in range(0,  6)])
    ax.set_yticks([i for i in range(-1, 2)])

    ax.set_title("The Gibbs phenomenon [N={}]".format(N), fontsize=20)
    ax.set_xlabel("$t$",  fontsize=20)
    ax.set_ylabel("$\hat x_N(t)$",  fontsize=20)

    plt.show()
```

### Fourier Series
$
\small{
\begin{equation}
   \begin{split}
    a_k = \frac{1}{T}\int_0^T x(t) e^{-jkw_0t} \,dt= \frac{1}{2}\int_0^1 1 e^{-jk\pi t} \,dt + \frac{1}{2}\int_1^2 -1 e^{-jk\pi t} \,dt
   \end{split}
   \end{equation}
}   
$


$
\small{
\begin{equation}
   \begin{split}
    \frac{1}{2}\int_0^1 \cos{(k \pi t)} - j \sin {(k \pi t)} \,dt - \frac{1}{2}\int_1^2 \cos{(k \pi t)} - j \sin {(k \pi t)} \,dt
   \end{split}
   \end{equation}
}   
$

$
\small{
\begin{equation}
   \begin{split}
    \frac{-j}{2}\int_0^1 \sin {(k \pi t)} \,dt - \frac{-j}{2}\int_1^2 \sin {(k \pi t)} \,dt = \frac{1}{2j}\int_0^1 \sin {(k \pi t)} \,dt - \frac{1}{2j}\int_1^2 \sin {(k \pi t)} \,dt = \frac{1}{2j k \pi} \left[ - \cos {(k \pi t)} \space \Big|_0^1 \right] - \frac{1}{2jk \pi} \left[ - \cos {(k \pi t)} \space \Big|_1^2 \right]
   \end{split}
   \end{equation}
}   
$

$
\small{
\begin{equation}
   \begin{split}
    \frac{-1}{2j k \pi} \left[ \cos {(k \pi t)} \space \Big|_0^1 \right] + \frac{1}{2jk \pi} \left[ \cos {(k \pi t)} \space \Big|_1^2 \right] = \frac{-1}{2j k \pi} \Big( \left[ {(-1)^k - 1} \right] - \left[ {1- (-1)^k} \right] \Big) &= \begin{cases}
                0 &\quad \space\space\space   k \space mod \space 2 = 0 \\
                \frac{2}{j k \pi} &\quad \space\space\space   k \space mod \space 2 \ne 0
        \end{cases}
   \end{split}
   \end{equation}
}   
$

$
\small{
\begin{equation}
   \begin{split}
       x(t) = \sum_{k=-\infty}^{+\infty} a_k e^{jkw_0t} = \sum_{k=odd \space numbers} \frac{2}{j k \pi} e^{jk \pi t} = \sum_{k=1}^{+\infty} \frac{2}{j (2k-1) \pi} \Big( e^{j(2k-1) \pi t} - e^{-j(2k-1) \pi t} \Big) = \sum_{k=1}^{+\infty} \frac{2}{j (2k-1) \pi} \Big(2j \sin ((2k-1) \pi t)\Big) = \sum_{k=1}^{+\infty} \frac{4 \sin \big ((2k-1) \pi t \big)}{ (2k-1) \pi} = \frac{4}{\pi}\sum_{k=1}^{+\infty} \frac{ \sin \big ((2k-1) \pi t \big )}{ 2k-1 }
   \end{split}
   \end{equation}
}   
$
### Pulse in Finite Sample Size
$
\small{
\begin{equation}
   \begin{split}
        \hat x_N(t) = \frac{4}{\pi} \sum_{k=1}^{+N} \frac{ \sin \big ((2k-1) \pi t \big)}{ 2k-1 }
   \end{split}
   \end{equation}
}   
$
