```python
import pandas as pd

# Question_3
# 3_a
#Sequence = pd.read_csv('Sequence.csv', ',', header=None, names=[ i for i in range(11)])
Sequence = open('Sequence.csv', 'r')
transactions = Sequence.read()
print(transactions)
```

```python
from gsppy.gsp import GSP
# 3_b
item_list = []
transactions_list = []
for line in transactions.split('\n'):
    item_list = []
    for item in line[:-1].split(','):
        item_list.append(item)
        
    transactions_list.append(item_list)
    
print(transactions_list)
```

```python

result = GSP(transactions_list).search(0.3)
```

```python
# 3_c
prev_items = []
for i in range(0, len(transactions_list)):
    if ('Bread' in transactions_list[i] and 
       'Sweet' in transactions_list[i] and 
        transactions_list[i].index('Sweet') - transactions_list[i].index('Bread') == 1):
            
        if transactions_list[i].index('Bread') != 0:
            for j in range(0 , transactions_list[i].index('Bread')):
                prev_items.append(transactions_list[i][j])
            
print('================== prev items ==================')
print(pd.Series(prev_items).value_counts())
```
