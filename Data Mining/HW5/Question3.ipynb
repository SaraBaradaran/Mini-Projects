```python
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram

# Question 3
# 3_a
iris = load_iris(as_frame=True)
x = iris.data
y = iris.target

display(x.head())
display(y.head())

Z = linkage(x, 'single')
```

```python
import matplotlib.pyplot as plt
# 3_b

fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z)
```

```python
from scipy.cluster.hierarchy import fcluster
# 3_c
#================== Using criterion='distance' ======================#
label = fcluster(Z, 6, criterion='distance')
print(label)
```

```python
# 3_d
#================== Using criterion='distance' ======================#
label = fcluster(Z, 0.5, criterion='distance')
print(label)
```

```python
#================== Using criterion='distance' ======================#
fig, axes = plt.subplots(1, 6, figsize=(20,5))
r = 0 
for i in [0.2, 0.4, 0.6, 0.7, 0.9, 1.8]:
    label = fcluster(Z, i, criterion='distance')
    axes[r].scatter(x['sepal length (cm)'],x['sepal width (cm)'],c=label, cmap=plt.cm.Set1)
    axes[r].set_title('Hierarchical : ' + str(i), fontsize=18)
    r += 1
```
