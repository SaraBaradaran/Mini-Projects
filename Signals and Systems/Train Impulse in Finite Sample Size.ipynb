```python
import numpy as np
import matplotlib.pyplot as plt

def Train_Impulse(N, T):
    
    x_t = np.linspace(-24, 24, 500)
    x_N = np.zeros_like(x_t)
    
    x_N = (np.sin((2*N+1) * np.pi * x_t / T)) / (T * np.sin(np.pi * x_t / T))

    plt.figure(figsize=(15, 5))
    plt.plot(x_t, x_N, 'b')
    plt.ylim(-1.2, 1.2)

    plt.xticks([i for i in range(-25,  26)])
    plt.yticks([i for i in range(-2, 6)])

    plt.title("The Train Impulse [N={}, T={}]".format(N, T), fontsize=20)
    plt.xlabel("$t$", fontsize=20)
    plt.ylabel("$\hat x_N(t)$", fontsize=20)
    
    A = ((2*N)+1)/T
    plt.axhline(A, color='y', linestyle='dashed')  
    
    sampleRange = int(np.ceil((25 / T)))
    Impulses = [i*T for i in range(-sampleRange+1, sampleRange)]
    labels = ['{}T'.format(i) for i in range(-sampleRange+1, sampleRange)]
    
    min_ylim, max_ylim = plt.ylim() 
    for i in range(len(Impulses)):
        plt.axvline(Impulses[i], color='y', linestyle='dashed')
        plt.text(Impulses[i], max_ylim * 0.8, labels[i], color='b', rotation=90, backgroundcolor='1')

    plt.show()
```

### Fourier Series
$
\small{
\begin{equation}
   \begin{split}
     a_k = \frac{1}{T}\int_0^T x(t) e^{-jkw_0t} \,dt= \frac{1}{T}\int_\frac{-T}{2}^\frac{T}{2} \big [\sum_{k=-\infty}^{+\infty} \delta (t-kT) \big ] e^{-jk\frac{2\pi}{T} t} \,dt = \frac{1}{T}
   \end{split}
   \end{equation}
}   
$


$
\small{
\begin{equation}
   \begin{split}
       x(t) = \sum_{k=-\infty}^{+\infty} a_k e^{jkw_0t} = \sum_{k=-\infty}^{+\infty} \frac{1}{T} e^{jk\frac{2\pi}{T}t}
   \end{split}
   \end{equation}
}   
$

### Train Impulse in Finite Sample Size

$
\small{
\begin{equation}
   \begin{split}
       \hat x_N(t) = \sum_{k=-N}^{+N} \frac{1}{T} e^{jk\frac{2\pi}{T}t} = \frac{1}{T} \sum_{k=-N}^{+N} (e^{j\frac{2\pi}{T}t})^k = \frac{1}{T} \frac{\sin \big( (2N+1)\frac{t\pi}{T} \big )}{\sin(\frac{t\pi}{T})} = \frac{\sin \big ((2N+1)\frac{t\pi}{T} \big )}{T\sin(\frac{t\pi}{T})}
   \end{split}
   \end{equation}
}   
$

### Domain Interpretation
$
\small{
\begin{equation}
   \begin{split}
                A = \lim_{t->0} \hat x_N(t) = \lim_{t->0} \big [ \frac{\sin \big ( (2N+1)\frac{t\pi}{T} \big )}{T\sin(\frac{t\pi}{T})} \big ] = \frac{(2N+1)\frac{\pi}{T} \space \cos \big ( (2N+1)\frac{t\pi}{T} \big )}{\frac{\pi}{T}T\cos(\frac{t\pi}{T})} = \frac{2N+1}{T}
   \end{split}
   \end{equation}
}   
$
