```python
import numpy as np
import matplotlib.pyplot as plt

def Synthesis_Approximation_of_Impulse_Signal(B, label):
    
    n = np.arange(-50, 50, 1);
    x = np.round(np.sin(B * n), 5) / (np.pi * n)

    x[50] = 1
    
    plt.figure(figsize=(15, 5))
    plt.yticks([i for i in range(-50, 50)])

    plt.title("Synthesis Approximation of Impulse Signal [B={}$\pi$]".format(label), fontsize=20)
    plt.xlabel("$n$", fontsize=20)
    plt.ylabel("$\hat x[n]$", fontsize=20)

    plt.stem(n, x);
```

### DTFT (Discrete Time Fourier Transform)

#### َAnalysis Equation 
$
\small{
\begin{equation}
   \begin{split}
     DTFT \{\delta [n]\} = \sum_{n=-\infty}^{+\infty} x [n]  e^{-jwn} = \sum_{n=-\infty}^{+\infty} \delta [n]  e^{-jwn} = 1
   \end{split}
   \end{equation}
}   
$

#### Synthesis Equation
$
\small{
\begin{equation}
   \begin{split}
       x[n] = \frac{1}{2\pi} \int_{-\infty}^{+\infty} 1 \times e^{jwn} dw = \delta [n]
   \end{split}
   \end{equation}
}   
$

### Synthesis Approximation

$
\small{
\begin{equation}
   \begin{split}
       \hat x[n] = \frac{1}{2\pi} \int_{-B}^{B} 1 \times e^{jwn} dw = \frac{\sin{Bn}}{\pi n}
   \end{split}
   \end{equation}
}   
$


```python
Synthesis_Approximation_of_Impulse_Signal(B=np.pi * 0.25, label="0.25")
```
